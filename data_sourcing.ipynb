```python
#We will be using the imdb non commercial dataset that can be found here: https://developer.imdb.com/non-commercial-datasets/
#We are only interested in hollywood movies going back 15 years.
```

```python
import pandas as pd


#title.basics.tsv
title_basics = pd.read_csv('title.basics.tsv', sep='\t')
title_basics.head()
```

```python
#title.ratings.tsv
title_ratings = pd.read_csv('title.ratings.tsv', sep = '\t')
title_ratings.head()
```

```python
#we will merge title_basics and title_ratings

title_main = title_basics.merge(title_ratings, how='left', on='tconst')
title_main.head()
```

```python
#title_main has a fair bit of information about the titles. It would be nice if we could include information like the production company, producer and director.
principals = pd.read_csv('title.principals.tsv', sep='\t')
names = pd.read_csv('name.basics.tsv', sep='\t')
principals_with_names = principals.merge(names,how='left',on='nconst')
print(principals_with_names)
```

```python
principals_with_names.head()
directors = principals_with_names[principals_with_names['category'] == 'director']
producers = principals_with_names[principals_with_names['category'] == 'producer']
```

```python
directors.head()
```

```python
main = title_main.merge(directors, how='left', on='tconst')
```

```python
#at this point, we have a lot of columns that are not important to us. we will drop those columns
main = main.drop(columns = ['isAdult','endYear','ordering','nconst','category','job','characters','birthYear','deathYear','primaryProfession','knownForTitles'])
```

```python
main.dtypes
```

```python
main = main.rename(columns = {'primaryName':'director'})
```

```python
producers.head()
producers = producers[['tconst','primaryName']]
```

```python
main = main.merge(producers,how = 'left',on='tconst')
main.head()
```

```python
main.rename(columns = {'primaryName':'producer'},inplace=True)
```

```python
main.head()
```

```python
main['titleType'].unique()
```

```python
main['startYear'].unique()
```

```python
main.dtypes
```

```python
import numpy as np

main['startYear'].replace('\\N', np.nan, inplace=True)
main['startYear'] = pd.to_numeric(main['startYear'], errors='coerce')

main.head()
```

```python
main['startYear'].replace(np.nan, 0, inplace=True)
```

```python
#convert runtime to numeric
main['runtimeMinutes'].replace('\\N', np.nan, inplace=True)
main['runtimeMinutes'].replace(np.nan, 0, inplace=True)
main['runtimeMinutes'] = pd.to_numeric(main['runtimeMinutes'], errors='coerce')
```

```python
main.head()
```

```python
#as we are only interested in movies going back 20 years (2024 - 2003), filter out the movies that are outside this range. Additionally, we will also filter out any records that are not movies
imdb_dataset = main[(main['startYear'] >= 2004) & (main['startYear'] <= 2024) & ((main['titleType'] == 'movie') | (main['titleType'] == 'tvMovie')) & (main['runtimeMinutes'] >= 60)]
imdb_dataset.head()
```

```python
imdb_dataset.tail()
```

```python
imdb_dataset.drop_duplicates(subset=['originalTitle'], inplace=True)
```
